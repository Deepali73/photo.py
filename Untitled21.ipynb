```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# 1. Create a blank image (RGB) - size 500x500 pixels
image = np.zeros((500, 500, 3), dtype=np.uint8)

# 2. Fill background with a light color (e.g., light blue)
image[:] = (255, 230, 200)  # BGR format

# 3. Draw shapes
# Draw a blue rectangle
cv2.rectangle(image, (50, 50), (450, 150), (255, 0, 0), thickness=-1)  # Filled rectangle

# Draw a green circle
cv2.circle(image, (250, 300), 75, (0, 255, 0), thickness=-1)  # Filled circle

# Draw a red line
cv2.line(image, (100, 400), (400, 400), (0, 0, 255), thickness=5)

# 4. Add text
cv2.putText(image, "Hello, OpenCV!", (70, 120), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

# 5. Display the image using matplotlib (for Jupyter compatibility)
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Generated Image from Scratch")
plt.axis('off')
plt.show()

# 6. (Optional) Save the image to disk
cv2.imwrite("created_image.jpg", image)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# 1. Create a blank image (500x300) with RGB channels
image = np.zeros((300, 500, 3), dtype=np.uint8)

# 2. Fill background with light pink
image[:] = (180, 200, 255)  # BGR (OpenCV uses Blue-Green-Red)

# 3. Add text "I"
cv2.putText(image, "I", (50, 200), cv2.FONT_HERSHEY_SIMPLEX, 4, (0, 0, 0), 8)

# 4. Draw a heart shape using a combination of circles and triangle (polygon)
# Coordinates for heart center
center_x, center_y = 200, 160

# Left circle
cv2.circle(image, (center_x - 25, center_y), 25, (0, 0, 255), -1)

# Right circle
cv2.circle(image, (center_x + 25, center_y), 25, (0, 0, 255), -1)

# Bottom triangle
points = np.array([[center_x - 50, center_y],
                   [center_x + 50, center_y],
                   [center_x, center_y + 70]], np.int32)
cv2.fillPoly(image, [points], (0, 0, 255))

# 5. Add text "YOU"
cv2.putText(image, "YOU", (270, 200), cv2.FONT_HERSHEY_SIMPLEX, 2.5, (0, 0, 0), 6)

# 6. Display using matplotlib (Jupyter-friendly)
plt.figure(figsize=(8, 5))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("I ❤️ YOU Image")
plt.axis('off')
plt.show()

# 7. (Optional) Save to disk
cv2.imwrite("i_love_you_image.jpg", image)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import random

# 1. Create a blank dark sky background
image = np.zeros((500, 800, 3), dtype=np.uint8)
image[:] = (10, 10, 30)  # Dark blue night sky

# 2. Add random stars (white dots)
for _ in range(300):
    x = random.randint(0, 799)
    y = random.randint(0, 499)
    radius = random.randint(1, 2)
    cv2.circle(image, (x, y), radius, (255, 255, 255), -1)

# 3. Draw the moon
cv2.circle(image, (700, 100), 40, (230, 230, 200), -1)  # Pale yellowish moon

# 4. Add silhouette of a person (head and body as circles/rectangles)
# Head
cv2.circle(image, (150, 400), 20, (0, 0, 0), -1)

# Body
cv2.rectangle(image, (140, 420), (160, 480), (0, 0, 0), -1)

# Optional: sitting legs
cv2.line(image, (160, 480), (180, 495), (0, 0, 0), 6)
cv2.line(image, (140, 480), (120, 495), (0, 0, 0), 6)

# 5. Add some stars as constellations (for effect)
constellation_points = [(400, 200), (420, 190), (440, 210), (460, 180), (480, 200)]
for pt in constellation_points:
    cv2.circle(image, pt, 2, (255, 255, 100), -1)
for i in range(len(constellation_points) - 1):
    cv2.line(image, constellation_points[i], constellation_points[i + 1], (255, 255, 100), 1)

# 6. Display using matplotlib (for Jupyter)
plt.figure(figsize=(12, 7))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Stargazing Scene 🌌")
plt.axis('off')
plt.show()

# 7. Optional: Save to disk
cv2.imwrite("stargazing_scene.jpg", image)
```
